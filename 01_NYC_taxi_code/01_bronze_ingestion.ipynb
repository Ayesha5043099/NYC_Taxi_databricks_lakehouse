```python
from pyspark.sql.functions import current_timestamp

df_bronze = df.withColumn("ingestion_time", current_timestamp())

df_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("nyc_taxi_project.raw_data.bronze_taxi")

print("Bronze table saved!")
```
